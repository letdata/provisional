```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from io import StringIO
import os
sns.set()
```

```python
scq_string = (
    'size,color,quantity\n'
    'L,Green,10\n'
    'S,Red,20\n'
    'M,White,5\n'
    'XS,White,0\n'
    'XL,Black,2\n'
    'L,Brown,0\n'
    'S,Yellow,15\n'
)
scq = pd.read_csv(StringIO(scq_string))
```

```python
scq['want'] = scq['quantity'].apply(lambda q: q % 4); scq.head(4)
```

```python
scq['M'] = scq.apply(lambda ROW: 'Yes' if ROW['quantity'] == ROW['want'] else 'No', axis='columns'); scq.head(4)
```
